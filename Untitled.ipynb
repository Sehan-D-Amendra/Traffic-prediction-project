# Data Preparation

```python
wt_features = ['Rain', 'Clouds', 'Clear', 'Snow', 'Mist',
              'Drizzle', 'Haze', 'Thunderstorm', 'Fog', 'Smoke', 'Squall']
wd_features = ['light rain', 'few clouds', 'Sky is Clear', 'light snow', 'sky is clear', 'mist', 'broken clouds', 'moderate rain', 'drizzle', 'overcast clouds', 'scattered clouds', 'haze', 'proximity thunderstorm', 'light intensity drizzle', 'heavy snow', 'heavy intensity rain', 'fog', 'heavy intensity drizzle', 'shower snow', 'snow', 'thunderstorm with rain',
              'thunderstorm with heavy rain', 'thunderstorm with light rain', 'proximity thunderstorm with rain', 'thunderstorm with drizzle', 'smoke', 'thunderstorm', 'proximity shower rain', 'very heavy rain', 'proximity thunderstorm with drizzle', 'light rain and snow', 'light intensity shower rain', 'SQUALLS', 'shower drizzle', 'thunderstorm with light drizzle']

new_weather_type_col = []
new_weather_description_col = []
for i in range(10000):
    if(weather_type_arr[i]) not in wt_features:
        new_weather_type_col.append(0)
    else:
        new_weather_type_col.append((wt_features.index(weather_type_arr[i]))+1)

    
    if weather_description_arr[i] not in wd_features:
        new_weather_description_col.append(0)
    else:
        new_weather_description_col.append((wd_features.index(weather_description_arr[i]))+1)
```

```python
data['weather_type'] = new_weather_type_col
```

```python
data['weather_description'] = new_weather_description_col
```

```python
features = numeric_columns+label_columns
target = ['traffic_volume']
X = data[features]
y = data[target]
```

```python
X.head(6)
```

```python
y.head(6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape , X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, Y_train)
    acc = model.score(X_test, Y_test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10, 50, 100],
              'criterion':['squared_error', 'absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf , param_grid=parameters)
grid_fit = grid_obj.fit(X_train, Y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test, Y_test)
```

```python
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```

# visualization

```python
metrics = ['month', 'month_day', 'weekday', 'hour']
fig = plt.figure(figsize=(8, 4*len(metrics)))
for i, metric in enumerate(metrics):
	ax = fig.add_subplot(len(metrics), 1, i+1)
	ax.plot(data.groupby(metric)['traffic_volume'].mean(), '-o')
	ax.set_xlabel(metric)
	ax.set_ylabel("Mean Traffic")
	ax.set_title(f"Traffic Trend by {metric}")
plt.tight_layout()
plt.show()
```
